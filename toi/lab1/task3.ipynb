```python
import numpy as np

S: np.float32
N: np.uint8

N = np.uint8(3)
S = np.float32(1.0) / N

iters = 0
while S != 1:
    print("iters =", iters, "S =", S)
    S = S + 1.0 / np.float32(N)
    iters += 1

print("iters =", iters, "S =", float(S))
```
