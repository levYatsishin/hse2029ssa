```python
import csv
import json

purchase_file = "purchase_log.txt"
visits_file = "visit_log.csv"
out_file = "funnel.csv"
```

```python
def load_purchase_categories(path: str) -> dict[str, str]:
    purchases: dict[str, str] = {}
    with open(path, encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)
            user_id: str = data["user_id"]
            category: str = data["category"]
            purchases[user_id] = category
    return purchases


def process_visits_stream(
    visits_path: str,
    purchases: dict[str, str],
    out_path: str,
) -> None:
    with open(visits_path, encoding="utf-8") as vin, \
         open(out_path, "w", encoding="utf-8", newline="") as vout:
        reader = csv.reader(vin)
        writer = csv.writer(vout)

        header: list[str] = next(reader)
        writer.writerow(["user_id", "source", "category"])

        for row in reader:
            user_id: str = row[0]
            source: str = row[1]
            if user_id in purchases:
                category: str = purchases[user_id]
                writer.writerow([user_id, source, category])
```

```python
purchases: dict[str, str] = load_purchase_categories(purchase_file)
process_visits_stream(visits_file, purchases, out_file)
```
