```python
# Исходные данные 
documents: list[dict[str, str]] = [
    {'type': 'passport', 'number': '2207 876234',
     'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2',
     'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006',
     'name': 'Аристарх Павлов'}
]

directories: dict[str, list[str]] = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def find_owner(doc_number: str, docs: list[dict[str, str]]) -> str | None:
    for doc in docs:
        if doc.get('number') == doc_number:
            return doc.get('name')
    return None


def find_shelf(doc_number: str, dirs: dict[str, list[str]]) -> str | None:
    for shelf, numbers in dirs.items():
        if doc_number in numbers:
            return shelf
    return None


def handle_command(cmd: str) -> None:
    # 'p' - владелец, 's' - полка.
    cmd = cmd.strip().lower()
    if cmd == 'p':
        num = input("Введите номер документа:\n").strip()
        owner = find_owner(num, documents)
        if owner is not None:
            print("Владелец документа:", owner)
        else:
            print("Документ не найден.")
        return

    if cmd == 's':
        num = input("Введите номер документа:\n").strip()
        shelf = find_shelf(num, directories)
        if shelf is not None:
            print("Документ хранится на полке:", shelf)
        else:
            print("Документ не найден.")
        return

    print("Неизвестная команда. Доступные: p, s, q.")
```

```python
def main() -> None:
    while True:
        cmd = input("Введите команду (p - владелец, s - полка, q - выход):\n")
        if cmd.strip().lower() == 'q':
            print("Выход.")
            break
        handle_command(cmd)
```

```python
main()
```
